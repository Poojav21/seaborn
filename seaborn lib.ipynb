# Plot a scatter plot between total_bill and tips

```python
# heatmaps--> generally plotted betwwn the two categorical data or one categorical or one discrete numerical d
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
x = flights.pivot_table(index= "year", columns = "month", values = "passengers", aggfunc = "sum")
x
```

```python
sns.heatmap(x, cbar = False)
plt.show()
```

```python
sns.heatmap(x, cbar = False, linewidth = 0.5)
plt.show()
```

```python
sns.heatmap(x, cbar = False, linewidth = 0.5, annot = True, fmt = "d", cmap = "viridis")
plt.show()
```

```python
sns.heatmap(x, cbar = False, linewidth = 0.5, annot = True, fmt = "d", cmap = "summer")
plt.show()
```

```python
sns.heatmap(x, cbar = False, linewidth = 0.5, annot = True, fmt = "d", cmap = "winter")
plt.show()
```

```python
# clustermaps
sns.clustermap(x)
plt.show()
```

```python
sns.clustermap(x, metric = "correlation")
plt.show()
```

```python
sns.clustermap(x, z_score = 0, annot = True, metric = "correlation")
plt.show()
```

```python
sns.clustermap(x, z_score = 0, annot = True, row_cluster = False, cmap = "winter", metric = "correlation")
plt.show()
```

```python
sns.clustermap(x, z_score = 0, annot = True, col_cluster = False, cmap = "winter", metric = "correlation")
plt.show()
```
